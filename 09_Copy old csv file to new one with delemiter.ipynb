```python
import csv

with open('data.csv', 'r') as file:
    csv_reader = csv.reader(file)
    
    with open('copy_data.txt', 'w') as copy_data:
        csv_writer = csv.writer(copy_data, delimiter="\t")
    
        for line in csv_reader:
            csv_writer.writerow(line)
            print(line[1] + " -- " + line[7])
```
